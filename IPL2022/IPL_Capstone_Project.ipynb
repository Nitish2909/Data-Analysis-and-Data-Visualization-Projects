Now, Here comes some Basic Questions

1. Which team won the most matches ?

```python
df['match_winner'].value_counts()
```

```python
match_wins=df['match_winner'].value_counts()
```

```python
# barplot for match wins
sns.barplot(y=match_wins.index,x=match_wins.values,palette="viridis")
plt.title("Most Matches Win By Team")
plt.xlabel("Number of Matches")
```

2. Toss Winner Trends(Means if won the toss what team decide)

```python
sns.countplot(x=df['toss_decision'],palette="flare")
plt.title("Toss Winner Trends")
plt.ylabel("Number of Toss")
plt.xlabel("Toss Winner Decision")
```

3. Toss Winner vs Match Winner

```python
# we check the percentage of (those team won the toss they also won match or not)
count= df[df['toss_winner'] == df['match_winner']]['match_id'].count()
percentage = (count*100)/df.shape[0]
print(percentage.round(2))
```

4. How team won by(means how many runs or Wicket team won their match) (Runs vs Wickets)

```python
sns.countplot(x=df['won_by'],palette="rainbow")
plt.title("Teams Won By How Many Runs and Wickets")
plt.ylabel("Times to Win Matches")
plt.xlabel("Won By Runs or Wickets")
```

**Key Player Performances**

1. Most "Player of the Match" Awards

```python
count = df['player_of_the_match'].value_counts().head(10)
sns.barplot(x= count.values, y=count.index, palette='mako')
plt.title("Top 10 Players That Won Most Player Of The Match Award")
plt.xlabel("Number of Player of The Match Awards")
plt.ylabel("Player Of The Match")
```

2. Top 2 Scorers

```python
high= df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
sns.barplot(high,palette="coolwarm")
plt.title("Top Scorers In IPL 2022")

```

3. Top 10 Best Bowling Figures

```python
df['Highest_wickets']= df['best_bowling_figure'].apply(lambda x:x.split("--")[0])
df["Highest_wickets"] = df["Highest_wickets"].astype(int)
top_bowlers = df.groupby('best_bowling')["Highest_wickets"].sum().sort_values(ascending=False).head(10)
top_bowlers.plot(kind="barh",color="orange")
plt.title("Highest Wicket Takers in IPL 2022")
plt.xlabel("Number of Wickets")
plt.ylabel("Name of the Highest Wickets Takers Bowlers")
```
